```python
import random

suits = ('Spades','Clover','Hearts','Diamond')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}
playing = True
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of '+ self.suit
```

```python
class Deck():
    #To create a new deck of cards
    def __init__ (self):
        self.all_cards =[]
        
        #To include 52 unique cards in the empty list created above
        for suit in suits:
            for rank in ranks:
                created_card = Card(suit,rank)
                #To add the newly created into the all cards list
                self.all_cards.append(created_card)
                
    def __str__(self):
        for card in self.all_cards:
            deck_comp += '\n'+ card.__str__()
            return "The deck has: "+deck_comp
                
    #To create a method to shuffle the deck that was properly created in an order
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    #To deal cards for the game
    def deal(self):
        one_card = self.all_cards.pop()
        return one_card
```

```python
class Hand():
    def __init__(self):
        self.cards =[]
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        
        if card.rank == 'Ace':
            self.aces += 1
        
    def adjust_for_aces(self):
        
        #If total value > 21 and player having an Ace in hand then change the total value
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips():
    def __init__(self,total = 100):
        self.total = total
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True :
        
        try:
            chips.bet = int (input("How many chips would you like to bet ? : "))
        except:
            print ("Sorry please provide a valid bet")
        else:
            if chips.bet > chips.total:
                print ("Sorry, You have insufficient chips! ")
            else:
                break
```

```python
def hit (deck,hand):
    
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_aces()
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    while True :
        x = input('Hit or Stand??')
        
        if x[0].lower() == 'h':
            hit(deck,hand)
            
        elif x[0].lower() == 's':
            print ("Player stands!! Dealer's turn" )
            playing =  False
            
        else:
            print ('Please enter either to Hit or Stand')
            continue
        break
```

```python
def show_some(player,dealer):
    
    #to show only one of dealer's card
    print ("\n Dealer's Hand: ")
    print ("First card is hidden")
    print(dealer.cards[1])
    
    #to show player's both cards
    print ("\n The Player's Hand: ")
    for card in player.cards:
        print(card)
    
```

```python
def show_all(player,dealer):
    
    #show all the dealer's cards
    print ("\n The Dealer's Hand: ")
    for card in dealer.cards:
        print(card)
        
    print(f"The value of Dealer's Hand is :{dealer.value}")
    
    #show all the player's cards
    print ("\n The Player's Hand: ")
    for card in player.cards:
        print(card)
        
    print(f"The value of Player's Hand is :{player.value}")
```
